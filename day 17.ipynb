```python
def KMPAlgo(P,S):
    M=len(P)
    N=len(S)
    lps=[]
    LPS(P,M,lps)
    print(lps)
    i=0
    j=0
    while (N-i)>=(M-j):
        if P[j]==S[i]:
            i+=1
            j+=1
            
        if j==M:
            print("pattern found",i-j)
            j=lps[j-1]
        elif i<N and P[j]!=S[i]:
            if j!=0:
                j=lps[j-1]
            else:
                i+=1
            
def LPS(P,M,lps): 
    lps.append(0)
    L=[0]
    j=0
    for i in range(1,M):
        if P[i]==P[j]:
            L.append(j+1)
            j=j+1
        else:
            j=0
            lps.append(j)


if __name__=="__main__":
    S="ABABACDEABABABCABCABCABCABDAA"
    P="ABCAB"
    KMPAlgo(P,S)
```

```python
ctr=[0]
def tower (n,frm,to,aux,ctr):
    if n==0:
        return
    tower(n-1,frm,aux,to,ctr)
    print(f"move {n} from {frm} to {to}")
    ctr [0]+=1
    tower (n+1,aux,to,frm,ctr)

n=4
ctr=[0]
tower(n,'A','C','B',ctr)
print(ctr)

    
```
